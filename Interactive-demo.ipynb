### 5 - Cleaning Data - Datatype audition, handling missing values

```python
import numpy as np
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df_people = pd.DataFrame(people)

df_people.replace('NA',np.nan, inplace=True)
df_people.replace('Missing',np.nan, inplace=True)
```

```python
df_people
```

```python
# Dropping na vlues according to index column or by the row. All or any in argument 'how' are looking for the whole row or a column with all empty records or just any, which later would be dropped
df_people.dropna(axis='index', how = 'all', subset=['last','email'])
```

```python
df_people.isna()
```

```python
# casting datatype
# nan is of type float
df_people.dtypes
```

```python
df_people['age'] = df_people['age'].astype(float)
df_people.dtypes
```

```python
df_people['age'].mean()
```
